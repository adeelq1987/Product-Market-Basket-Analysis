# **Customer Market Basket Analysis**

## Business Problem

The Sales and Marketing leadership has asked if there is buying pattern for our customers.  By creating the market basket analysis the hope is to come up with a way to target products and discounts to customers to drive revenue.  

We will be using the rules based association model to understand the patterns.



```python
# Import important libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing Data to Google Colab
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['Hos_data.csv']))
```

```python
df1 = df.set_index('Invoice_No')
```

```python
def encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1
```

```python
basket_encode_plus = df1.applymap(encode_units)
basket_encode_plus
```

```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.03, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
```

```python
frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
product_analysis = association_rules(frequent_itemsets_plus, metric='lift', min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```

```python
df_prod = pd.DataFrame(product_analysis)
```

```python
df_prod.head()
```

The support column tell us what percentage of the product mix applied to antecedents and consequents.  It is a good way for marketing personnel to understand what the customers buying pattern is.
